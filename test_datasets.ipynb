```python
import os
import numpy as np
from matplotlib import pyplot as plt
from datasets.mnist_variants import *
img_rows, img_cols = 8, 8
```

```python
colored_mnist = ColoredMNIST(
    root=os.path.join('.', 'downloads', 'ColoredMNIST'),
    train=True,
    download=True,
    spurious_correlations=[0.1, 0.2, 0.9]
)
indices = np.random.permutation(range(len(colored_mnist)))[:img_rows*img_cols]
fig, axes = plt.subplots(img_rows, img_cols, figsize=(4*img_cols, 4*img_rows))
for row_idx in range(img_rows):
    for col_idx in range(img_cols):
        ax = axes[row_idx, col_idx]
        image, label, mdata = colored_mnist[indices[img_rows*row_idx+col_idx]]
        image = torch.cat((image, torch.zeros_like(image[0, :, :]).unsqueeze(0)), dim=0).permute(1, 2, 0)
        image = 0.5*image+0.5
        ax.imshow(image)
        ax.set_title('Label: {}, Env: {}'.format(label, mdata['environment']))
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
rotated_mnist = RotatedMNIST(
    root=os.path.join('.', 'downloads', 'RotatedMNIST'),
    train=True,
    download=True,
    rotations=[0, 15, 30, 45, 60, 75]
)
indices = np.random.permutation(range(len(rotated_mnist)))[:img_rows*img_cols]
fig, axes = plt.subplots(img_rows, img_cols, figsize=(4*img_cols, 4*img_rows))
for row_idx in range(img_rows):
    for col_idx in range(img_cols):
        ax = axes[row_idx, col_idx]
        image, label, mdata = rotated_mnist[indices[img_rows*row_idx+col_idx]]
        image = image.permute(1, 2, 0)
        image = 0.5*image+0.5
        ax.imshow(image, cmap='binary')
        ax.set_title('Label: {}, Env: {}'.format(label, mdata['environment']))
        ax.set_xticks([])
        ax.set_yticks([])
```
